#### Qual o comportamento dos outliers?
Usando boxplot

```python
# criando lista de colunas numéricas
num_cols = [x for x in data.columns if x not in categ_cols+['default','participacao_falencia_valor']]
```

```python
fig, axs = plt.subplots(nrows=int(len(num_cols)/2), ncols=2, figsize=(10, 26))
fig.suptitle("Boxplot to look for outliers")

for feat, ax in zip(num_cols, axs.ravel()):
    sns.boxplot(y=data[feat], ax=ax, palette="Blues")
```

Para as colunas "month" e "year" é importante observarmos se não há nenhuma anotação estranha - se por exemplo "month" tivesse um valor 13 saberíamos que há algo errado.

Para as demais features, observamos que a grande maioria dos valores são considerados outliers pela regra do interquartile range (IQR) - assim como esperado, já que como vimso no começo 75% dos valores eram 0. O que pode ser eliminado aqui são valores tão distantes que mudam a ordem de grandeza, como é o caso de "quant_protestos" e "dividas_vencidas_valor".

Para os demais, pela grandeza dos valores, pode ser feita a normalização ou operação sobre a feature

## Correlação
Um dos maiores indicadores de relevância de uma feature é a relação que ela tem com o target

```python
# para correlação não linear, Spearman
corr = data[num_cols+["default"]].corr(method="spearman")
plt.subplots(figsize=(15,10))
sns.heatmap(corr, xticklabels=corr.columns,yticklabels=corr.columns, annot=True, )
```

#### Como muda a correlação se aplicamos uma operação logarítmica?

```python
# avaliando correlação com o log das features - indicativo de o que fazer no feature engineering
log_data = data[num_cols].apply(np.log)
log_data['default'] = data["default"]
log_corr = log_data.corr(method="spearman")
plt.subplots(figsize=(15,10))
sns.heatmap(log_corr, xticklabels=log_corr.columns,yticklabels=log_corr.columns, annot=True, )
```

```python
corr["delta_log"] = abs(log_corr["default"])-abs(corr["default"])
corr["delta_log"]
```

Observamos um ganho considerável na correlação se usamos o logaritmo das colunas "valor_protestos", "acao_judicial_valor", "dividas_vencidas_qtd", podemos incluir essa operação no tratamento de dados.

```python
# obersvando mais de perto a relação com 'default_3months' já que é a variável mais correlata

sns.kdeplot(data=data, x='default_3months', hue='default', fill=True)

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Default 3 months')
plt.ylabel('Densidade de default')
```

Quanto mais defaults nos últimos 3 meses, maior é a quantidade de defaults atuais

#### Podemos eliminar algumas features com baixa correlação

```python
# threshold sugerido para eliminar features que não estão relacionadas
thr_corr=0.01
best_feats = corr[abs(corr["default"]) > thr_corr].index
best_feats
```

Baseados na correlação, podemos cortar algumas features não relacionadas com o target. Além disso, temos uma ideia das features que são redundantes - i.e. muito correlacionadas entre si - como "quant_protestos" e "valor_protestos", "dividas_vencidas_valor" e "dividas_vencidas_quant". Podemos combiná-las ou usar apenas uma delas.

#### Testes com PCA (principal component analysis)
Uma forma de combinar as variáveis e reduzir dimensionalidade.

```python
# "quant_protestos" e "valor_protestos", muito correlacionadas
pca = PCA(n_components=1)
data['pca_protestos'] = pca.fit_transform(data[['quant_protestos', 'valor_protestos']])
data.default.corr(data.pca_protestos, method="spearman")
```

A correlação não tem ganho fazendo esse processo, então podemos usar diretamente 'valor_protestos' (corr=0.044)

```python
data['pca_divida'] = pca.fit_transform(data[['dividas_vencidas_qtd', 'dividas_vencidas_valor']])
data.default.corr(data.pca_divida, method="spearman")
```

A correlação teve perda, então podemos usar diretamente 'dividas_vencidas_qtd' (corr=-0.012)

Vamos fazer o mesmo teste para as variáveis de tempo.

```python
data['pca_date'] = pca.fit_transform(data[['month', 'year']])
data.default.corr(data.pca_date, method="spearman")
```

Resultado ligeiramente melhor que 'month' (corr=0.085). Pode ser mais interessante esse tratamento do que o uso da coluna anteriormente criada "date" por não passar por um encoder e perder informação.

# Conclusão

Principais insights:
- O problema se trata de um dataset desbalanceado
- O default tem uma tendência temporal, concentrada no começo de cada ano e podemos considerar uma feature que combina mês e ano usando PCA
- As colunas com mais correlação com o target são: ['default_3months', 'ioi_36months', 'ioi_3months', 'valor_vencido', 'valor_quitado', 'quant_protestos', 'valor_protestos', 'quant_acao_judicial', 'acao_judicial_valor', 'dividas_vencidas_qtd', 'valor_total_pedido', 'month', 'year']
- As colunas numéricas estão altamente concentradas no zero, então nossa informação pode vir justamente dos valores por vezes considerados outliers. Além disso as grandezas tem influência na predição dependendo do modelo utilizado, por isso operações como logaritmo podem influenciar a correlação com o target.
- As colunas "atividade_principal" e "forma_pagamento" não poderão ser tratadas com OneHotEncoder pois isso tornará nossos dados esparsos, nem LabelEncoder pois valores únicos do teste podem não existir no treino por ter uma alta granularidade.
